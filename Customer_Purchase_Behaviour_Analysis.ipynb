```python
import pandas as pd
```

```python
df=pd.read_excel("Online Retail.xlsx")
df
```

```python
df.info()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df['Description']=df['Description'].fillna('Unknown')
```

```python
df.info()
```

```python
df['StockCode']=df['StockCode'].str.strip()
```

```python
df['Isreturn']=df['Quantity'].apply(lambda x:1 if x<0 else 0)
```

```python
df['Quantity']=df['Quantity'].apply(lambda x:0 if x<0 else x)
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df['UnitPrice']=df['UnitPrice'].apply(lambda x:0 if x<0 else x)
```

```python
df['Revenue']=df['Quantity']*df['UnitPrice']
```

```python
df = df.dropna(subset=['CustomerID'])
```

```python
df['CustomerID'].nunique()
```

```python
df['InvoiceNo'].count()
```

```python
df['Revenue'].sum()
```

```python
AOV = df.groupby('InvoiceNo')['Revenue'].sum().mean()
AOV
```

```python
df['CustomerID']=df['CustomerID'].astype(int)
```

Top 10 Customers by revenue

```python
freq_customers = df.groupby('CustomerID')['InvoiceNo'].nunique().sort_values(ascending=False).head(10)
freq_customers
```

```python
spend = df.groupby('CustomerID')['Revenue'].sum()
spend.describe()
```
